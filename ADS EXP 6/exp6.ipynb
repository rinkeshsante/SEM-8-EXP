```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb
```

```python

df = pd.read_csv("../Toyota.csv", index_col=0, na_values=["??","???"])
df_copy = df.copy()
df.dropna(axis=0, inplace=True)
df.head()
```

```python
# scatterplot

plt.subplot(1,2,1)
plt.scatter(df["Age"],df["Price"])
plt.title("Age vs Price")


plt.subplot(1,2,2)
plt.scatter(df["KM"],df["Price"])
plt.title("KM vs Title")
```

```python
# histogram

plt.subplot(1,2,1)
plt.hist(df["Age"])
plt.title("Age")

plt.subplot(1,2,2)
plt.hist(df["Price"])
plt.title("Price")
```

```python
# bar charts

count = df["FuelType"].value_counts()
count = dict(count)

plt.bar(*zip(*count.items()), color = ['red', 'blue', "gray"])
```

```python
# box and Whiskers with matplotlib

plt.boxplot(df["Age"])
```

```python
# box with whisker in seaborn

sb.boxplot(x=df["FuelType"],y=df["Price"])
```

```python
# pairwise plot

sb.pairplot(df, kind="scatter")
```
